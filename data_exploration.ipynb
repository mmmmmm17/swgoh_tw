```python
def get_ally_codes(ally_code):
    """
    returns a list with all ally codes from player's guild
    
    ally_code: ally code of a player; integer
    """
    guild = client.get_data('guild', ally_code)
    guild_members = guild[0]['roster']
    # empty list
    ally_codes = list()
    for i in range(len(guild_members)):
        ally_codes.append(guild_members[i]['allyCode'])
    return ally_codes
```

```python
def how_many_toons(ally_codes, stars, time_sleep = 50):
    """
    returns: sorted list from most uncommon to most common characters in a guild
    
    A character must be unlocked by at least a single person to be present in a returned list.
    
    ally_codes: ally codes of players in a guild; could be any iterable
    stars: at least how many stars; integer
    time_sleep: timeout between requests; integer
    """
    # empty dict 
    toons = dict()
    for ally in ally_codes:
        units = client.get_data('units', ally)
        # there is an authorization error without timeout between requests
        time.sleep(time_sleep)
        for unit in units.keys():
            if unit in toons and units[unit][0]['starLevel'] >= 6:
                toons[unit] += 1
            elif units[unit][0]['starLevel'] >= 6:
                toons[unit] = 1
            elif unit not in toons:
                toons[unit] = 0
            else:
                pass
    return sorted(toons.items(), key=lambda x: x[1])
```
